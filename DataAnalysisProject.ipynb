```python
import pandas as pd
import matplotlib.pyplot as mlpp
import numpy as np
import statistics as stat
from collections import Counter
from tabulate import tabulate

dataset1 = pd.read_csv("HPAP001.pooled.csv")
dataset2 = pd.read_csv("HPAP003.csv")
dataset3 = pd.read_csv("HPAP004.csv")
dataset4 = pd.read_csv("HPAP015.csv")
dataset5 = pd.read_csv("HPAP030.csv")
dataset6 = pd.read_csv("HPAP031.csv")
```

```python
#Question 1: To find the correaltion between clone size and mutation:
#Alternative Hypothesis:Are avg_V_identity and copies strongly correlated (correlation coefficnet > 0.6) 

#Accuring V_idnetity from all datasets
v_identity1 = dataset1['avg_v_identity']
v_identity2 = dataset2['avg_v_identity']
v_identity3 = dataset3['avg_v_identity']
v_identity4 = dataset4['avg_v_identity']
v_identity5 = dataset5['avg_v_identity']
v_identity6 = dataset6['avg_v_identity']
v_identity = []
for x in range(0,len(v_identity1)-1):
    v_identity.append(v_identity1[x])
for x in range(0,len(v_identity2)-1):
    v_identity.append(v_identity2[x])
for x in range(0,len(v_identity3)-1):
    v_identity.append(v_identity3[x])
for x in range(0,len(v_identity4)-1):
    v_identity.append(v_identity4[x])
for x in range(0,len(v_identity5)-1):
    v_identity.append(v_identity5[x])
for x in range(0,len(v_identity6)-1):
    v_identity.append(v_identity6[x])

#Accuring copies from all datasets
copies1 = dataset1['copies']
copies2 = dataset2['copies']
copies3 = dataset3['copies']
copies4 = dataset4['copies']
copies5 = dataset5['copies']
copies6 = dataset6['copies']
copies = []
for x in range(0,len(copies1)-1):
    copies.append(copies1[x])
for x in range(0,len(copies2)-1):
    copies.append(copies2[x])
for x in range(0,len(copies3)-1):
    copies.append(copies3[x])
for x in range(0,len(copies4)-1):
    copies.append(copies4[x])
for x in range(0,len(copies5)-1):
    copies.append(copies5[x])
for x in range(0,len(copies6)-1):
    copies.append(copies6[x])

#Find the covaraince for avg_V_identity and copies
covariacne  = np.cov(v_identity,copies)
covariacne = covariacne[0][1]

#Find the mean and var for V_identity and copies
avg_v_identity = stat.mean(v_identity)
avg_copies = stat.mean(copies)
var_v_identity =stat.variance(v_identity)
var_copies = stat.variance(copies)

#Setting up values for a table
Head = ["----","Mean","Varaince"]
V_identity = ["V_identity",avg_v_identity,var_v_identity]
Copies = ["Copies",avg_copies,var_copies]
merged = [Head,V_identity,Copies]
table = tabulate(merged,tablefmt = "fancy_grid")
print(table)

#Find the correlation for avg
correlation = (covariacne)/((var_v_identity**(1/2))*(var_copies)**(1/2))
print("Correlation:",correlation)

#A correlation value of 0.06 was obtained indicating that avg_V_identity and copies are not correlated at all. 
```
